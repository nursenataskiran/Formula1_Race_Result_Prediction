```python
from final_cleaner import F1DataCleaner
from LSTMDataPrep import LSTMDataPreparer
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Path to the folder containing the PKL files
folder_path = "./all_f1_data" 

# Create an instance of the cleaner class
cleaner = F1DataCleaner(folder_path)

# Start the cleaning process for all files in the folder
cleaner.process_files()
```

```python
cleaned_sessions = cleaner.cleaned_sessions
```

```python
preparer = LSTMDataPreparer(cleaned_sessions)
x_data, y_data, scaler, test_sessions = preparer.prepare()
```

```python
y_data.train
```

```python
x_data.test.shape
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, TimeDistributed, LSTM, Dense

input_layer = Input(shape=(20, 40, 9))
x = TimeDistributed(LSTM(64))(input_layer)
output = TimeDistributed(Dense(20, activation='softmax'))(x)

model = Model(inputs=input_layer, outputs=output)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(
    x_data.train, y_data.train, 
    validation_data=(x_data.test, y_data.test),
    batch_size=32,
    epochs=50
)
```

```python
# Loss
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title("Loss Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

# Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Accuracy Over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
y_pred = model.predict(x_data.test)
```

```python
# Get the class with the highest probability (position prediction)
y_pred_labels = np.argmax(y_pred, axis=-1)  # shape: (num_races, 20)

# True labels (actual race positions)
y_true = y_data.test  # shape: (num_races, 20)

# Example: print the real and predicted rankings for the first 3 races
for i in range(3):
    print(f"Race {i+1}")
    print("Real   :", y_true[i])
    print("Predicted:", y_pred_labels[i])
    print("---")
```

```python
correct = (y_pred_labels == y_true)
accuracy = correct.sum() / correct.size
print(f"Total Accuracy: {accuracy:.4f}")
```

```python
# Get predictions
y_pred = model.predict(x_data.test)
y_pred_labels = np.argmax(y_pred, axis=-1)

# List to store results
results = []

# Process results for each race
for race_idx, session in enumerate(test_sessions):
    # Race information
    race_date = session.session_info['StartDate'].date()
    race_name = session.session_info['Meeting']['Name']
    race_info = f"{race_date} – {race_name}"

    # Sort drivers by abbreviation
    results_sorted = session.results.sort_values("Abbreviation").reset_index(drop=True)
    drivers = results_sorted['Abbreviation'].values

    # True and predicted positions for this race
    y_true_race = y_data.test[race_idx]
    y_pred_race = y_pred_labels[race_idx]

    # Create a row in the table for each driver
    for i, driver_abbr in enumerate(drivers):
        results.append({
            "Race Info": race_info,
            "Driver": driver_abbr,
            "True Position": y_true_race[i] +1,
            "Predicted Position": y_pred_race[i] +1,
            "Correct": y_true_race[i] == y_pred_race[i]
        })

# Create a DataFrame
df_results = pd.DataFrame(results)
```

```python
from sklearn.metrics import mean_absolute_error

# Flatten y_true and y_pred into vectors
y_true_all = y_data.test.reshape(-1)
y_pred_all = y_pred_labels.reshape(-1)

# Calculate MAE
mae = mean_absolute_error(y_true_all, y_pred_all)
print(f"Mean Absolute Error (MAE): {mae:.4f}")
```

```python
# Convert predicted probability matrices into unique race position assignments 
# using the Hungarian algorithm (ensures one driver ↔ one position mapping)

y_pred_labels = []

for race_probs in y_pred:  # race_probs shape: (20, 20)
    from scipy.optimize import linear_sum_assignment

    # Convert probabilities to a cost matrix (to minimize)
    cost_matrix = -race_probs  

    # Find the optimal one-to-one assignment between drivers and positions
    row_ind, col_ind = linear_sum_assignment(cost_matrix)

    # Build label array: driver[row_ind[i]] → position[col_ind[i]]
    labels = np.zeros(20, dtype=int)
    labels[row_ind] = col_ind

    y_pred_labels.append(labels)

y_pred_labels = np.array(y_pred_labels)
```

```python
results = []

for race_idx, session in enumerate(test_sessions):
    # Actual race information
    race_date = session.session_info['StartDate'].date()
    race_name = session.session_info['Meeting']['Name']
    race_info = f"{race_date} – {race_name}"

    drivers = session.results['Abbreviation'].values
    y_true_race = y_data.test[race_idx]
    y_pred_race = y_pred_labels[race_idx]

    for i, driver_abbr in enumerate(drivers):
        results.append({
            "Race Info": race_info,
            "Driver": driver_abbr,
            "True Position": y_true_race[i],
            "Predicted Position": y_pred_race[i],
            "Correct": y_true_race[i] == y_pred_race[i]
        })

# Convert the results into a DataFrame
df_results = pd.DataFrame(results)
```
